```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
measurements = Path('measurements')
```

```python
t = [x for x in measurements.glob("*.csv")][0].parts[-1]
```

```python
def get_pdu_and_itvl(filename):
    t = filename
    pdu = t[t.find('_') + 1: t.find('_', t.find('_') + 1)]
    itvl = t[t.rfind('_') + 1: t.rfind('.csv')]
    return int(pdu), int(itvl)
```

```python
d = {}
for x in measurements.glob("*.csv"):
    df = pd.read_csv(x)
    pdu, itvl = get_pdu_and_itvl(x.parts[-1])
    
    if not pdu in d.keys():
        d[pdu] = {}
    
    d[pdu][itvl] = df
```

```python
d[1][1000]
```

```python
mean_current = pd.DataFrame(columns=['pdu', 'itvl', 'mean', 'max', 'min', 'std', 'median'])
for pdu in d.keys():
    for itvl, df in d[pdu].items():
        cur = {
            'pdu': pdu,
            'itvl': itvl,
            'mean': df['Current (uA)'].mean(),
            'max': df['Current (uA)'].max(),
            'min': df['Current (uA)'].min(),
            'std': df['Current (uA)'].std(),
            'median': df['Current (uA)'].median(),
              }
        mean_current = mean_current.append(cur, ignore_index=True)

mean_current.sort_values(['pdu', 'itvl'], inplace=True)
```

```python
pdus = sorted(mean_current['pdu'].unique())
n_pdus = len(pdus)
last_pdu = None

fig, ax = plt.subplots(nrows=n_pdus// 2 + (n_pdus % 2), ncols = 2, figsize=(16, 2 * n_pdus))
fig_all, ax_all = plt.subplots(figsize=(20, 8))
ax=ax.ravel()
for e, pdu in enumerate(pdus):
    mean_current.loc[mean_current['pdu'] == pdu, ['mean', 'itvl']].set_index('itvl').plot(ax=ax[e])
    mean_current.loc[mean_current['pdu'] == pdu, ['mean', 'itvl']].set_index('itvl').plot(ax=ax_all, label=pdu)
    if last_pdu is not None:
        ax[e].scatter(x=[100], y=last_pdu, color='r')

        
    ax[e].set_title(pdu)
    ax[e].set_ylabel("Current uA")
    ax[e].set_xlabel("Advertisement Interval (ms)")
    ax[e].grid()
    ax[e].legend([pdu])

        
    last_pdu = mean_current.loc[(mean_current['pdu'] == pdu) & (mean_current['itvl'] == 100), 'mean']
    
ax_all.set_title('All in one')
ax_all.set_ylabel("Current uA")
ax_all.set_xlabel("Advertisement Interval (ms)")
ax_all.grid()
ax_all.legend(pdus)
fig.tight_layout()
```

```python
itvls = sorted(mean_current['itvl'].unique())
n_itvls = len(itvls)

fig, ax = plt.subplots(nrows=n_itvls // 2 + n_itvls % 2, ncols = 2, figsize=(16, 2 * n_itvls))
fig_all, ax_all = plt.subplots(figsize=(20, 8))
ax=ax.ravel()
for e, itvl in enumerate(itvls):
    mean_current.loc[mean_current['itvl'] == itvl, ['mean', 'pdu']].set_index('pdu').plot(ax=ax[e])
    mean_current.loc[mean_current['itvl'] == itvl, ['mean', 'pdu']].set_index('pdu').plot(ax=ax_all, label=pdu)

    ax[e].set_title(itvl)
    ax[e].set_ylabel("Current uA")
    ax[e].set_xlabel("PDU Length (Bytes)")
    ax[e].grid()
    ax[e].legend([itvl])

    
ax_all.set_title('All in one')
ax_all.set_ylabel("Current uA")
ax_all.set_xlabel("PDU Length (Bytes)")
ax_all.grid()
ax_all.legend(itvls)
fig.tight_layout()
```
